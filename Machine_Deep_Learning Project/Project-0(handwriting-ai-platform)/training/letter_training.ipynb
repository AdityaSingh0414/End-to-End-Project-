# EMNIST Handwritten Letter Recognition Model (CRNN)

This notebook demonstrates how to design and train a Convolutional Recurrent Neural Network (CRNN) to read handwritten letters. CRNN combines convolutional layers (for visual features) with recurrent LSTM layers (to evaluate sequential context).

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

## 1. CRNN Model Architecture

The CRNN processes letter slices (28x28x1). After applying convolutional and max-pooling filters, it flattens features along the vertical height and runs a Bidirectional LSTM over the horizontal coordinates, classifying characters into one of 26 alphabet labels (A-Z).

```python
def build_letter_crnn():
    model = models.Sequential([
        layers.Input(shape=(28, 28, 1)),
        # CNN Layers
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)), # output shape: (14, 14, 32)
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)), # output shape: (7, 7, 64)
        
        # Reshape: Treats the height spatial dimension as sequence elements
        # Reshaping from (7, 7, 64) -> (7, 448) (7 sequence steps, each with 448 features)
        layers.Reshape((7, 448)),
        
        # Recurrent Layer (Bidirectional LSTM)
        layers.Bidirectional(layers.LSTM(64, return_sequences=False)),
        
        # Classification Head
        layers.Dense(64, activation='relu'),
        layers.Dense(26, activation='softmax')
    ])
    
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

model = build_letter_crnn()
model.summary()
```
