# MNIST Handwritten Digit Recognition Model

This notebook covers the step-by-step implementation for building, training, and evaluating the CNN model used for handwritten digit recognition.

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

## 1. Load and Preprocess the Dataset

We use Keras's built-in MNIST dataset containing 60,000 training and 10,000 test images.

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to [0, 1]
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

print(f"Train shape: {x_train.shape}")
print(f"Test shape: {x_test.shape}")
```

## 2. Define CNN Architecture

```python
model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

## 3. Train Model

```python
history = model.fit(x_train, y_train, epochs=5, batch_size=128, validation_data=(x_test, y_test))
```

## 4. Save Trained Model Weights

```python
os.makedirs('../models', exist_ok=True)
model.save('../models/cnn_digit_model.h5')
print("Digit recognition model saved successfully.")
```
