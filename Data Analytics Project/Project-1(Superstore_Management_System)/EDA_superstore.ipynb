UNVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt

print("Categorical Variables Analysis:\n")

categorical_cols = [
    'Customer Segment',
    'Category',
    'Region',
    'State',
    'City',
    'Payment Mode',
    'Delivery Status',
    'Auto Reorder'
]

for col in categorical_cols:
    print(f"----- {col} -----")

    # Value counts
    counts = df[col].value_counts()
    percentages = df[col].value_counts(normalize=True) * 100

    print(counts)
    print("\nPercentage Distribution:")
    print(percentages.round(2))
    print("\n")

    # Handle large categories (like City)
    if df[col].nunique() > 10:
        counts = counts.head(10)

    # Visualization
    plt.figure(figsize=(8,4))
    counts.plot(kind='bar')

    plt.title(f"Top Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

print("Numerical Variables Analysis:\n")

# Correct column names based on your dataset
numerical_cols = ['Sales Amount', 'Quantity', 'Discount (%)', 'Profit']

for col in numerical_cols:
    print(f"----- {col} -----")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")
    print(f"Minimum: {df[col].min():.2f}")
    print(f"Maximum: {df[col].max():.2f}")
    print(f"25th Percentile: {df[col].quantile(0.25):.2f}")
    print(f"50th Percentile: {df[col].quantile(0.50):.2f}")
    print(f"75th Percentile: {df[col].quantile(0.75):.2f}")
    print("\n")

    # Visualization
    plt.figure(figsize=(10,4))

    # Histogram
    plt.subplot(1,2,1)
    df[col].hist(bins=30, alpha=0.7)
    plt.title(f"{col} Distribution")
    plt.xlabel(col)
    plt.ylabel("Frequency")

    # Boxplot
    plt.subplot(1,2,2)
    df.boxplot(column=col, vert=False)
    plt.title(f"{col} Boxplot")

    plt.tight_layout()
    plt.show()
```

```python
#Bivariate Analysis

import matplotlib.pyplot as plt

print("Bivariate Analysis:\n")

# FIXED: include Order ID count
category_performance = df.groupby('Category').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'count'   # ✅ added this line
}).rename(columns={'Order ID': 'Total Count'})

# Profit Margin
category_performance['Profit Margin (%)'] = (
    category_performance['Profit'] / category_performance['Sales Amount']
) * 100

print(category_performance, "\n")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Sales Amount
category_performance['Sales Amount'].plot(
    kind='bar', ax=axes[0, 0], color='skyblue'
)
axes[0, 0].set_title('Total Sales Amount by Category')
axes[0, 0].tick_params(axis='x', rotation=45)

# Profit
category_performance['Profit'].plot(
    kind='bar', ax=axes[0, 1], color='salmon'
)
axes[0, 1].set_title('Total Profit by Category')
axes[0, 1].tick_params(axis='x', rotation=45)

# Profit Margin
category_performance['Profit Margin (%)'].plot(
    kind='bar', ax=axes[1, 0], color='lightgreen'
)
axes[1, 0].set_title('Profit Margin (%) by Category')
axes[1, 0].tick_params(axis='x', rotation=45)

# Order Count (NOW WORKS ✅)
category_performance['Total Count'].plot(
    kind='bar', ax=axes[1, 1], color='purple', alpha=0.7
)
axes[1, 1].set_title('Total Order Count by Category')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Regional Performance Analysis

print("Regional Performance Analysis:\n")
region_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'count'   # ✅ added this line
}).rename(columns={'Order ID': 'Total Count'})

region_performance['Profit Margin (%)'] = (
    region_performance['Profit'] / region_performance['Sales Amount'])*100
print(region_performance, "\n")

#Visualization
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
region_performance['Sales Amount'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Amount Distribution by Region')

plt.subplot(2, 2, 2)
region_performance['Profit'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Profit Distribution by Region')

plt.subplot(2, 2, 3)
sns.barplot(x=region_performance.index, y='Total Count', data=region_performance)
plt.title('Order Count by Region')
plt.xlabel('Region')
plt.ylabel('Order Count')

plt.subplot(2, 2, 4)
sns.barplot(x=region_performance.index, y='Profit Margin (%)', data=region_performance)
plt.title('Profit Margin (%) by Region')
plt.xlabel('Region')
plt.ylabel('Profit Margin (%)')

plt.tight_layout()
plt.show()
```

```python
# TIME SERIES ANALYSIS

#Extract year and month from Order Date
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.to_period('M')

#Monthly Sales Trend
monthly_sales = df.groupby('Month').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'count'   # ✅ added this line
}).rename(columns={'Order ID': 'Total Count'})

print("Monthly Sales Trend:\n")
print(monthly_sales.tail(12), "\n")

#Visualization
plt.figure(figsize=(14, 6))

plt.subplot(3, 1, 1)
monthly_sales['Sales Amount'].plot(kind='line', marker='o')
plt.title('Monthly Sales Amount Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid(True)


plt.subplot(3, 1, 2)
monthly_sales['Profit'].plot(kind='line', marker='o', color='orange')
plt.title('Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.grid(True)

plt.subplot(3, 1, 3)
monthly_sales['Total Count'].plot(kind='line', marker='o', color='green')
plt.title('Monthly Order Count Trend')
plt.xlabel('Month')
plt.ylabel('Order Count')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
#ADVANCED ANALYSIS

#select numerical columns for correlation analysis
numerical_cols = ['Quantity', 'Unit Price','Discount (%)', 'Sales Amount', 'Cost Price', 'Profit']

#Correlation Matrix
correlation_matrix = df[numerical_cols].corr()
print("Correlation Matrix:\n")
print(correlation_matrix, "\n")

#heatmap visualization
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()

```

```python
# Customer Segmentation Analysis

# Customer Segmentation Analysis

segment_analysis = df.groupby('Customer Segment').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum',
    'Order ID': 'count'
}).round(2)

segment_analysis.columns = ['Total Sales', 'Total Profit', 'Total Quantity', 'Total Orders']

# ✅ ADD THESE LINES (FIX)
segment_analysis['Avg Profit'] = segment_analysis['Total Profit'] / segment_analysis['Total Orders']
segment_analysis['Avg Discount'] = df.groupby('Customer Segment')['Discount (%)'].mean()

print("Customer Segment Analysis:\n")
print(segment_analysis, "\n")

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Total Sales
segment_analysis['Total Sales'].plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Total Sales by Customer Segment')
axes[0, 0].tick_params(axis='x', rotation=45)

# Total Profit
segment_analysis['Total Profit'].plot(kind='bar', ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Total Profit by Customer Segment')
axes[0, 1].tick_params(axis='x', rotation=45)

# Avg Profit ✅
segment_analysis['Avg Profit'].plot(kind='bar', ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Average Profit per Order')
axes[1, 0].tick_params(axis='x', rotation=45)

# Avg Discount ✅
segment_analysis['Avg Discount'].plot(kind='bar', ax=axes[1, 1], color='purple', alpha=0.7)
axes[1, 1].set_title('Average Discount by Segment')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# KEY INSIGHTS  AND SUMMARY

print("Key Insights and Summary:\n")

#Overall Metrics
total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_per_order = (total_profit / total_orders)*100

print(f"Overall Performance:")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
print(f"Average Profit per Order: ${avg_profit_per_order:.2f}\n")

#Top Performing Categories
top_category = df.groupby('Category')['Profit'].sum().sort_values(ascending=False).index[0]
print(f"Top Performing Category: {top_category}\n")
for i,(category, profit) in enumerate(df.groupby('Category')['Profit'].sum().sort_values(ascending=False).items(), 1):
    print(f"{i}. {category}: ${profit:,.2f}")


#Regional Performance
best_region = df.groupby('Region')['Profit'].sum().sort_values(ascending=False).index[0]
print(f"\nBest Performing Region: {best_region}\n")

#Delivery Performance
delivery_stats= df['Delivery Status'].value_counts(normalize=True)*100
print("Delivery Performance:")
for status, percentage in delivery_stats.items():
    print(f"{status}: {percentage:.2f}%")


#Customer Segments
best_segment = df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\nMost Profitable Customer Segment: {best_segment}\n")
```

```python
# Flatten results into structured list
results_list = [
    ['Total Sales', total_sales],
    ['Total Profit', total_profit],
    ['Total Orders', total_orders],
    ['Average Profit per Order', avg_profit_per_order],
    ['Top Performing Category', top_category],
    ['Best Region', best_region],
    ['Most Profitable Customer Segment', best_segment]
]

# Add Delivery Performance separately
for status, count in delivery_stats.items():
    results_list.append([f'Delivery - {status}', count])

# Convert to DataFrame
results_df = pd.DataFrame(results_list, columns=['Metric', 'Value'])

# Export
results_df.to_csv('superstore_analysis_results.csv', index=False)

print("✅ Analysis results exported successfully!")
```
